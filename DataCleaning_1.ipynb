```python
#import modules
import os
import glob
import pandas as pd
import numpy as np
#list all csv files only
csv_files = glob.glob('*.{}'.format('csv'))
csv_files
```

```python
df_append = pd.DataFrame()
#append all files together
for file in csv_files:
            df_temp = pd.read_csv(file)
            df_append = df_append.append(df_temp, ignore_index=True)
df_append.head()
```

```python
df_append.info()
```

```python
print ("Rows     : " ,df_append.shape[0])
print ("Columns  : " ,df_append.shape[1])
print ("\nFeatures : \n" ,df_append.columns.tolist())
print ("\nUnique values :  \n",df_append.nunique())
```

```python
#Checking for the NULL values

df_append.isnull().sum()
```

```python
#Dropping the columns

df_append.drop(['Efficient use of medical imaging national comparison footnote','HCAHPS Answer Percent Footnote','Timeliness of care national comparison footnote','Effectiveness of care national comparison footnote','Patient experience national comparison footnote','Readmission national comparison footnote','Safety of care national comparison footnote','Mortality national comparison footnote','Hospital overall rating footnote','Survey Response Rate Percent Footnote','Number of Completed Surveys Footnote','Patient Survey Star Rating Footnote'],axis=1, inplace=True)
```

```python
#Checking for the Null Values
df_append.isnull().sum()
```

```python
#Replacing the Blank values for the column (Meets criteria for promoting interoperability of EHRs) with N

df_append[['Meets criteria for promoting interoperability of EHRs']] = df_append[['Meets criteria for promoting interoperability of EHRs']].fillna('N')
```

```python
#Checking for the Null Values
df_append.isnull().sum()
```

```python
#Dropping the rows of the County names where ever blank values were there

df_append.dropna( how='any',subset=['County Name'], axis=0,inplace=True)
```

```python
#Checking for the Null Values
df_append.isnull().sum()
```

```python
print ("\nUnique values :  \n",df_append.nunique())
```
