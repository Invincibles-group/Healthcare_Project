```python
df2 = ds.replace('Not Available',np.nan)
```

```python
df2 = df2.replace('Not Applicable',np.nan)
```

```python
df2['Count_Not_Available']=df2.isnull().sum(axis=1)
#df3['Count_Not_Applicable']=df3.isnull().sum(axis=1)
```

```python
df2['Total']=df2['Count_Not_Available']# + df3['Count_Not_Applicable']
```

```python
rslt_df = df2[df2['Total'] < 1]
```
